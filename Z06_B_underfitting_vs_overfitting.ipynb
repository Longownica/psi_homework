# Underfitting vs. Overfitting

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn import  metrics
```

Rozważmy następujący zbiór punktów:

```python
np.random.seed(0)
n_samples = 30
true_fun = lambda X: np.cos(1.5 * np.pi * X)
X = np.sort(np.random.rand(n_samples))
y = true_fun(X) + np.random.randn(n_samples) * 0.1
```

```python
# Plot outputs
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,1])
axes.set_ylim([-1.5,1.5])
plt.scatter(X, y,  color='black')
x_tr = np.linspace(0, 1, 200)
plt.show()
```

Podziel dane na dwie grupy.

```python
s=np.random.random_sample(n_samples)
s[s>0.5]=1
s[s<=0.5]=0
s
X1=X[s==1]
y1=y[s==1]
X2=X[s==0]
y2=y[s==0]

# Plot outputs
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,1])
axes.set_ylim([-1.5,1.5])
plt.scatter(X1, y1,  color='blue')
plt.scatter(X2, y2,  color='red')
x_tr = np.linspace(0, 1, 200)
plt.show()
```

# Zadanie
Wykonaj regresję:

* liniową
* wielomianową stopnia 20

w każdej z grup.

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn import  metrics
from sklearn import datasets, linear_model
```

```python
model1 = linear_model.LinearRegression()
model1.fit(np.vstack(X1), y1)

model20 = make_pipeline(PolynomialFeatures(20), linear_model.LinearRegression())
model20.fit(np.vstack(X2), y2)
```

```python
plt.figure(figsize=(6,6));
axes = plt.gca()
axes.set_xlim([0,1])
axes.set_ylim([-1.5,1.5])
plt.scatter(X1, y1,  color='blue')
plt.scatter(X2, y2,  color='red')
plt.plot(np.vstack(X1), model1.predict(np.vstack(X1)), color='blue',linewidth=3)
plt.plot(np.vstack(X2), model20.predict(np.vstack(X2)), color='red',linewidth=3)
x_tr = np.linspace(0, 1, 200)
plt.show()
```

# Zadanie
Wykonaj 100 podziałów na dwie równe części. Za każdym razem naucz regresję:

* liniową
* wielomianową stopnia 1, 2, 3, 4, 5, 6

na jednym fragmęcie i policz R_square na drugim.
Zwizualizuj wynik.


```python
for _ in range(1,100):
    s=np.random.random_sample(n_samples)
    s[s>0.5]=1
    s[s<=0.5]=0
    X1=X[s==1]
    y1=y[s==1]
    X2=X[s==0]
    y2=y[s==0]
    
    model1 = linear_model.LinearRegression()
    model1.fit(np.vstack(X1), y1)

    model2 = make_pipeline(PolynomialFeatures(2), linear_model.LinearRegression())
    model2.fit(np.vstack(X1), y1)
    
    model3 = make_pipeline(PolynomialFeatures(3), linear_model.LinearRegression())
    model3.fit(np.vstack(X1), y1)
    
    model4 = make_pipeline(PolynomialFeatures(4), linear_model.LinearRegression())
    model4.fit(np.vstack(X1), y1)
    
    model5 = make_pipeline(PolynomialFeatures(5), linear_model.LinearRegression())
    model5.fit(np.vstack(X1), y1)
    
    model6 = make_pipeline(PolynomialFeatures(6), linear_model.LinearRegression())
    model6.fit(np.vstack(X1), y1)
    
    plt.figure(figsize=(6,6));
    axes = plt.gca()
    axes.set_xlim([0,1])
    axes.set_ylim([-1.5,1.5])
    plt.scatter(X1, y1,  color='blue')
    plt.scatter(X2, y2,  color='red')
    plt.plot(np.vstack(X1), model1.predict(np.vstack(X1)), color='blue',linewidth=3)
    plt.plot(np.vstack(X1), model2.predict(np.vstack(X1)), color='red',linewidth=3)
    plt.plot(np.vstack(X1), model3.predict(np.vstack(X1)), color='yellow',linewidth=3)
    plt.plot(np.vstack(X1), model4.predict(np.vstack(X1)), color='green',linewidth=3)
    plt.plot(np.vstack(X1), model5.predict(np.vstack(X1)), color='purple',linewidth=3)
    plt.plot(np.vstack(X1), model6.predict(np.vstack(X1)), color='orange',linewidth=3)
    x_tr = np.linspace(0, 1, 200)
    plt.show()
    
    model22 = make_pipeline(PolynomialFeatures(2), linear_model.LinearRegression())
    model22.fit(np.vstack(X2), y2)
    print("R^2: {}".format(metrics.r2_score(y2, model22.predict(np.vstack(X2)))))
    
    
```
